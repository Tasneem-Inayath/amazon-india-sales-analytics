```python
# question 1: Revenue Trend with Visualization
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def eda_revenue_trend(cleaned_folder, eda_folder):
    """
    Generate yearly revenue trend, percentage growth, and save CSV + PNG visualization.
    Works on all cleaned datasets in cleaned_folder.
    """
    os.makedirs(eda_folder, exist_ok=True)
    
    # Gather all CSV files in cleaned folder
    all_files = glob.glob(os.path.join(cleaned_folder, "*.csv"))
    df_list = []
    
    for file in all_files:
        df = pd.read_csv(file)
        # Ensure numeric
        df['final_amount_inr'] = pd.to_numeric(df['final_amount_inr'], errors='coerce').fillna(0)
        df_list.append(df[['order_year', 'final_amount_inr']])
    
    # Combine all datasets
    combined_df = pd.concat(df_list)
    
    # Aggregate yearly revenue
    revenue_df = combined_df.groupby('order_year')['final_amount_inr'].sum().reset_index()
    
    # Calculate percentage growth
    revenue_df['pct_growth'] = revenue_df['final_amount_inr'].pct_change() * 100
    revenue_df['pct_growth'] = revenue_df['pct_growth'].fillna(0)
    
    # Save CSV
    csv_file = os.path.join(eda_folder, "eda_revenue_trend.csv")
    revenue_df.to_csv(csv_file, index=False)
    print(f"Revenue trend CSV saved to: {csv_file}")
    
    # ---- Visualization ----
    plt.figure(figsize=(12,6))
    sns.lineplot(data=revenue_df, x='order_year', y='final_amount_inr', marker='o', label='Revenue')
    
    # Annotate percentage growth
    for i, row in revenue_df.iterrows():
        plt.text(row['order_year'], row['final_amount_inr']*1.01, f"{row['pct_growth']:.1f}%", 
                 horizontalalignment='center', fontsize=9, color='blue')
    
    plt.title("Amazon India Yearly Revenue Trend (2015-2025)")
    plt.xlabel("Year")
    plt.ylabel("Revenue (INR)")
    plt.grid(True)
    
    # Highlight key growth periods
    max_rev_year = revenue_df.loc[revenue_df['final_amount_inr'].idxmax(), 'order_year']
    max_rev_val = revenue_df['final_amount_inr'].max()
    plt.axvline(max_rev_year, color='red', linestyle='--', alpha=0.5)
    plt.text(max_rev_year, max_rev_val*1.05, f"Peak Revenue {int(max_rev_val):,}", color='red')
    
    # Save PNG
    png_file = os.path.join(eda_folder, "eda_revenue_trend.png")
    plt.tight_layout()
    plt.savefig(png_file)
    plt.close()
    
    print(f"Revenue trend plot saved to: {png_file}")
    return revenue_df
```

```python
# question 2:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def eda_seasonal_sales(dataset_path, eda_folder):
    """
    Analyze seasonal patterns: monthly revenue trends and category comparison.
    Saves summary CSV and heatmap PNG in the eda folder.
    """
    os.makedirs(eda_folder, exist_ok=True)
    
    # Load dataset
    df = pd.read_csv(dataset_path)
    
    # Ensure order_date is datetime
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    
    # Extract year and month
    df['order_year'] = df['order_date'].dt.year
    df['order_month'] = df['order_date'].dt.month
    df['month_name'] = df['order_date'].dt.strftime('%b')
    
    # 1️⃣ Monthly revenue across all categories
    monthly_sales = df.groupby(['order_year','order_month']).agg(
        revenue=('final_amount_inr','sum'),
        orders=('transaction_id','count')
    ).reset_index()
    
    # Save CSV
    csv_file = os.path.join(eda_folder, f"seasonal_sales_summary_{os.path.basename(dataset_path)}")
    monthly_sales.to_csv(csv_file, index=False)
    
    # Heatmap: all revenue
    pivot_all = monthly_sales.pivot(index='order_year', columns='order_month', values='revenue')
    plt.figure(figsize=(12,6))
    sns.heatmap(pivot_all, annot=True, fmt=".0f", cmap="YlGnBu")
    plt.title("Monthly Revenue Heatmap (All Categories)")
    plt.xlabel("Month")
    plt.ylabel("Year")
    plt.xticks(ticks=range(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
    png_file_all = os.path.join(eda_folder, f"seasonal_sales_heatmap_all_{os.path.basename(dataset_path).replace('.csv','.png')}")
    plt.savefig(png_file_all, bbox_inches='tight')
    plt.close()
    
    # 2️⃣ Category-wise seasonal heatmap
    category_monthly = df.groupby(['order_year','order_month','category']).agg(
        revenue=('final_amount_inr','sum')
    ).reset_index()
    
    # Pivot for heatmap (years as rows, months as columns) per category
    for cat in category_monthly['category'].unique():
        cat_data = category_monthly[category_monthly['category']==cat].pivot(index='order_year', columns='order_month', values='revenue')
        plt.figure(figsize=(12,6))
        sns.heatmap(cat_data, annot=True, fmt=".0f", cmap="coolwarm")
        plt.title(f"Monthly Revenue Heatmap ({cat})")
        plt.xlabel("Month")
        plt.ylabel("Year")
        plt.xticks(ticks=range(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
        png_file_cat = os.path.join(eda_folder, f"seasonal_sales_heatmap_{cat}_{os.path.basename(dataset_path).replace('.csv','.png')}")
        plt.savefig(png_file_cat, bbox_inches='tight')
        plt.close()
    
    print(f"Seasonal EDA saved: CSV -> {csv_file}, Heatmaps -> {png_file_all} + category PNGs")
    
    return monthly_sales
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def perform_rfm_analysis(dataset_path,output_folder):
    os.makedirs(output_folder, exist_ok=True)
    df = pd.read_csv(dataset_path)
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    df = df.dropna(subset=['order_date','customer_id','final_amount_inr'])
    
    reference_date = df['order_date'].max() + pd.Timedelta(days=1)
    
    rfm = df.groupby('customer_id').agg({
        'order_date': lambda x: (reference_date - x.max()).days,
        'transaction_id': 'nunique',
        'final_amount_inr': 'sum'

    }).reset_index()

    rfm.columns = ['CustomerID','Recency','Frequency','Monetary']

    rfm['R_Score'] = pd.qcut(rfm['Recency'],5,labels=[5,4,3,2,1])
    rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
    rfm['M_Score'] = pd.qcut(rfm['Monetary'],5,labels=[5,4,3,2,1])

    rfm['RFM_Score']=rfm[['R_Score','F_Score','M_Score']].astype(int).sum(axis=1)

    def segment_customer(score):
        if score>=12:
            return 'Loyal'
        elif score>=9:
            return 'Potential Loyalist'
        elif score>=6:
            return 'Need Attention'
        else:
            return 'At Risk'
    rfm['Segment'] = rfm['RFM_Score'].apply(segment_customer)
    
    # --- Step 6: Visualizations ---
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Segment', palette='viridis')
    plt.title('RFM Segmentation: Recency vs Monetary')
    plt.savefig(os.path.join(output_folder, 'rfm_recency_monetary.png'))
    plt.close()

    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Segment', palette='coolwarm')
    plt.title('RFM Segmentation: Frequency vs Monetary')
    plt.savefig(os.path.join(output_folder, 'rfm_frequency_monetary.png'))
    plt.close()

    # --- Step 7: Save Results ---
    rfm.to_csv(os.path.join(output_folder, 'rfm_analysis_results.csv'), index=False)

    print(f"✅ RFM analysis completed! Results saved in '{output_folder}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def eda_payment_trend(data_path, output_folder):
    """
    Analyze and visualize payment method trends (2015–2025) using stacked area chart.
    Shows rise of UPI and decline of COD over time.
    """

    # Create output folder
    os.makedirs(output_folder, exist_ok=True)

    # Load dataset
    df = pd.read_csv(data_path)

    # --- Step 1: Clean & Prepare Data ---
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    df = df.dropna(subset=['order_date', 'payment_method', 'final_amount_inr'])

    # Extract year
    df['year'] = df['order_date'].dt.year

    # Filter for 2015–2025
    df = df[(df['year'] >= 2015) & (df['year'] <= 2025)]

    # --- Step 2: Aggregate total revenue by year and payment method ---
    payment_trend = df.groupby(['year', 'payment_method'])['final_amount_inr'].sum().reset_index()

    # --- Step 3: Pivot for stacked area chart ---
    payment_pivot = payment_trend.pivot(index='year', columns='payment_method', values='final_amount_inr').fillna(0)

    # --- Step 4: Calculate % share (optional for insights) ---
    payment_pivot_pct = payment_pivot.div(payment_pivot.sum(axis=1), axis=0) * 100

    # --- Step 5: Plot stacked area chart (absolute values) ---
    plt.figure(figsize=(10, 6))
    payment_pivot.plot.area(ax=plt.gca(), cmap='tab20', alpha=0.85)
    plt.title('Evolution of Payment Methods (2015–2025)', fontsize=14)
    plt.xlabel('Year')
    plt.ylabel('Total Revenue (INR)')
    plt.legend(title='Payment Method', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'payment_trend_absolute.png'))
    plt.close()

    # --- Step 6: Plot stacked area chart (% share) ---
    plt.figure(figsize=(10, 6))
    payment_pivot_pct.plot.area(ax=plt.gca(), cmap='tab20', alpha=0.9)
    plt.title('Payment Method Market Share (%) Over Time (2015–2025)', fontsize=14)
    plt.xlabel('Year')
    plt.ylabel('Market Share (%)')
    plt.legend(title='Payment Method', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'payment_trend_share.png'))
    plt.close()

    # --- Step 7: Highlight key insights ---
    summary = payment_pivot_pct.iloc[[-1]].T.sort_values(by=payment_pivot_pct.index[-1], ascending=False)
    summary.to_csv(os.path.join(output_folder, 'payment_trend_summary.csv'))

    print(f"✅ Payment trend analysis complete! Results saved in '{output_folder}'")


```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
# question 5:
def eda_subcategory_performance_combined(sales_path, catalog_path, output_folder):
    """
    Perform subcategory-wise revenue analysis and save combined results:
    - Aggregate revenue per subcategory per year
    - Calculate YoY growth
    - Calculate market share
    - Save a single CSV and visualizations (bar, pie, treemap)
    """
    os.makedirs(output_folder, exist_ok=True)
    
    # --- Load datasets ---
    df_sales = pd.read_csv(sales_path, usecols=['product_id','final_amount_inr','order_year'])
    df_catalog = pd.read_csv(catalog_path, usecols=['product_id','subcategory'])
    
    # --- Merge to get subcategory ---
    df = df_sales.merge(df_catalog, on='product_id', how='left')
    df.dropna(subset=['final_amount_inr', 'subcategory', 'order_year'], inplace=True)
    
    # --- Aggregate revenue per subcategory per year ---
    subcat_revenue = df.groupby(['subcategory','order_year'])['final_amount_inr'].sum().reset_index()
    
    # --- Calculate YoY growth ---
    subcat_revenue['pct_growth'] = subcat_revenue.groupby('subcategory')['final_amount_inr'].pct_change()*100
    
    # --- Calculate total revenue and market share ---
    total_revenue = subcat_revenue.groupby('subcategory')['final_amount_inr'].sum().reset_index()
    total_revenue.rename(columns={'final_amount_inr':'total_revenue'}, inplace=True)
    total_revenue['market_share_pct'] = total_revenue['total_revenue'] / total_revenue['total_revenue'].sum() * 100
    
    # --- Merge back total revenue & market share with yearly revenue ---
    combined_df = subcat_revenue.merge(total_revenue, on='subcategory', how='left')
    
    # --- Save single CSV ---
    combined_csv_path = os.path.join(output_folder, 'subcategory_revenue_analysis.csv')
    combined_df.to_csv(combined_csv_path, index=False)
    
    # --- Visualizations ---
    
    # Bar chart: total revenue per subcategory
    plt.figure(figsize=(12,6))
    sns.barplot(x='subcategory', y='total_revenue', data=total_revenue.sort_values('total_revenue', ascending=False))
    plt.xticks(rotation=45, ha='right')
    plt.title("Total Revenue per Subcategory")
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'subcategory_revenue_bar.png'))
    plt.close()
    
    # Pie chart: market share
    plt.figure(figsize=(8,8))
    plt.pie(total_revenue['total_revenue'], labels=total_revenue['subcategory'], autopct='%1.1f%%', startangle=140)
    plt.title("Market Share by Subcategory")
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'subcategory_market_share_pie.png'))
    plt.close()
    
    # Treemap: total revenue
    plt.figure(figsize=(12,6))
    squarify.plot(sizes=total_revenue['total_revenue'], label=total_revenue['subcategory'], alpha=0.8)
    plt.axis('off')
    plt.title("Revenue Treemap by Subcategory")
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'subcategory_revenue_treemap.png'))
    plt.close()
    
    print(f"Subcategory analysis completed! CSV and visualizations saved in {output_folder}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# question 6:
def eda_prime_impact_analysis(
    sales_path: str,
    catalog_path: str,
    output_folder: str
):
    """
    Analyze impact of Prime membership on customer behavior.
    Compares:
        - Average order value (AOV)
        - Order frequency per customer
        - Subcategory preferences
    Generates multiple visualizations and saves them to output_folder.
    """

    # Step 1: Load data
    sales = pd.read_csv(sales_path)
    catalog = pd.read_csv(catalog_path)

    # Step 2: Merge with catalog for subcategory info
    df = pd.merge(
        sales[['customer_id', 'final_amount_inr', 'product_id', 'is_prime_member']],
        catalog[['product_id', 'subcategory']],
        on='product_id',
        how='left'
    )

    # Step 3: Clean data
    df.dropna(subset=['final_amount_inr'], inplace=True)
    df['is_prime_member'] = df['is_prime_member'].map({True: 'Prime', False: 'Non-Prime'})

    # Step 4: Create output folder
    os.makedirs(output_folder, exist_ok=True)

    # Step 5: --- Average Order Value (AOV) ---
    aov = df.groupby('is_prime_member')['final_amount_inr'].mean().reset_index()
    plt.figure(figsize=(6,4))
    sns.barplot(data=aov, x='is_prime_member', y='final_amount_inr', palette='coolwarm')
    plt.title("Average Order Value: Prime vs Non-Prime")
    plt.xlabel("")
    plt.ylabel("Average Order Value (INR)")
    plt.tight_layout()
    plt.savefig(f"{output_folder}/average_order_value.png")
    plt.close()

    # Step 6: --- Order Frequency per Customer ---
    order_freq = df.groupby(['is_prime_member', 'customer_id']).size().reset_index(name='order_count')
    freq_summary = order_freq.groupby('is_prime_member')['order_count'].mean().reset_index()

    plt.figure(figsize=(6,4))
    sns.barplot(data=freq_summary, x='is_prime_member', y='order_count', palette='viridis')
    plt.title("Average Order Frequency: Prime vs Non-Prime")
    plt.xlabel("")
    plt.ylabel("Orders per Customer")
    plt.tight_layout()
    plt.savefig(f"{output_folder}/order_frequency.png")
    plt.close()

    # Step 7: --- Subcategory Preferences ---
    subcat_pref = df.groupby(['is_prime_member', 'subcategory']).agg({
        'final_amount_inr': 'sum'
    }).reset_index()

    plt.figure(figsize=(12,6))
    sns.barplot(
        data=subcat_pref,
        x='subcategory',
        y='final_amount_inr',
        hue='is_prime_member',
        palette='Set2'
    )
    plt.title("Subcategory Revenue Comparison: Prime vs Non-Prime")
    plt.xticks(rotation=75)
    plt.ylabel("Revenue (INR)")
    plt.tight_layout()
    plt.savefig(f"{output_folder}/subcategory_revenue_comparison.png")
    plt.close()

    # Step 8: --- Market Share Pie Chart ---
    total_revenue = df.groupby('is_prime_member')['final_amount_inr'].sum()
    plt.figure(figsize=(5,5))
    total_revenue.plot(
        kind='pie',
        autopct='%1.1f%%',
        colors=['#66c2a5', '#fc8d62']
    )
    plt.title("Revenue Share: Prime vs Non-Prime")
    plt.ylabel("")
    plt.tight_layout()
    plt.savefig(f"{output_folder}/revenue_share_pie.png")
    plt.close()

    # Step 9: Save summaries
    aov.to_csv(f"{output_folder}/average_order_value_summary.csv", index=False)
    freq_summary.to_csv(f"{output_folder}/order_frequency_summary.csv", index=False)
    subcat_pref.to_csv(f"{output_folder}/subcategory_performance.csv", index=False)

    print(f"✅ Prime impact analysis completed. Results saved in: {output_folder}")
```

```python
import pandas as pd
import plotly.express as px
from geopy.geocoders import Nominatim
import time
import os
import json
# question 7:
def geographic_sales_analysis_amazon(df, save_path="output", geojson_path=r"C:\Projects\amazonanalytics\geojson\india_states.geojson"):
    """
    Geographic sales analysis for Amazon India dataset.
    Saves:
        - Summary CSVs
        - Tier and city charts as PNG
        - State choropleth as interactive HTML
    Does NOT return any figures.
    
    Parameters:
        df (DataFrame): Amazon transactions data
        save_path (str): Folder to save all charts and summaries
        geojson_path (str): Path to India states GeoJSON
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # --- Step 1: Revenue ---
    df['revenue'] = df['final_amount_inr']
    
    # --- Step 2: Aggregate data ---
    state_revenue = df.groupby('customer_state')['revenue'].sum().reset_index()
    tier_revenue = df.groupby('customer_tier')['revenue'].sum().reset_index()
    
    df['order_year'] = pd.to_datetime(df['order_date'], errors='coerce').dt.year
    tier_growth = df.groupby(['customer_tier', 'order_year'])['revenue'].sum().reset_index()
    
    # --- Step 3: Load GeoJSON ---
    with open(geojson_path, "r", encoding="utf-8") as f:
        india_geo = json.load(f)

    # --- Step 4: Map dataset states to GeoJSON states ---
    geojson_states = [feature['properties']['NAME_1'] for feature in india_geo['features']]
    state_revenue['customer_state'] = state_revenue['customer_state'].apply(
        lambda x: x if x in geojson_states else x.replace("&", "and").strip()
    )
    
    # --- Step 5: State choropleth ---
    fig_state = px.choropleth(
        state_revenue,
        geojson=india_geo,
        locations='customer_state',       # must match NAME_1 in GeoJSON
        featureidkey='properties.NAME_1',
        color='revenue',
        hover_name='customer_state',
        color_continuous_scale='YlOrRd',
        title='Revenue by Indian State'
    )
    fig_state.update_geos(fitbounds="locations", visible=False)
    fig_state.write_html(f"{save_path}/state_revenue_map.html")
    
    # --- Step 6: City map ---
    unique_cities = df[['customer_city', 'customer_state']].drop_duplicates()
    geolocator = Nominatim(user_agent="amazon_geo")
    latitudes, longitudes = [], []

    for _, row in unique_cities.iterrows():
        try:
            location = geolocator.geocode(f"{row['customer_city']}, {row['customer_state']}, India")
            if location:
                latitudes.append(location.latitude)
                longitudes.append(location.longitude)
            else:
                latitudes.append(None)
                longitudes.append(None)
            time.sleep(0.1)
        except:
            latitudes.append(None)
            longitudes.append(None)

    unique_cities['latitude'] = latitudes
    unique_cities['longitude'] = longitudes
    
    city_revenue = df.groupby('customer_city')['revenue'].sum().reset_index()
    city_revenue = city_revenue.merge(unique_cities[['customer_city', 'latitude', 'longitude']], on='customer_city', how='left')
    
    fig_city = px.scatter_mapbox(
        city_revenue,
        lat='latitude',
        lon='longitude',
        size='revenue',
        color='revenue',
        color_continuous_scale='Viridis',
        size_max=30,
        zoom=4,
        mapbox_style="carto-positron",
        hover_name='customer_city',
        title='Revenue by City'
    )
    fig_city.write_image(f"{save_path}/city_revenue.png")
    
    # --- Step 7: Tier bar chart ---
    fig_tier = px.bar(
        tier_revenue,
        x='customer_tier',
        y='revenue',
        color='customer_tier',
        title='Revenue by Tier'
    )
    fig_tier.write_image(f"{save_path}/tier_revenue.png")
    
    # --- Step 8: Tier growth line chart ---
    fig_growth = px.line(
        tier_growth,
        x='order_year',
        y='revenue',
        color='customer_tier',
        markers=True,
        title='Revenue Growth Patterns by Tier'
    )
    fig_growth.write_image(f"{save_path}/tier_growth.png")
    
    # --- Step 9: Save summaries ---
    state_revenue.to_csv(f"{save_path}/state_revenue_summary.csv", index=False)
    tier_revenue.to_csv(f"{save_path}/tier_revenue_summary.csv", index=False)
    tier_growth.to_csv(f"{save_path}/tier_growth_summary.csv", index=False)

    print(f"All charts and summaries saved in '{save_path}' folder.")
```

```python
import pandas as pd
import plotly.express as px
import os
# question 8:
def festival_sales_analysis_individual_plots(df, save_path="output"):
    """
    Festival sales analysis with a separate PNG for each festival.
    Saves:
        - Combined CSV of all festivals
        - Individual PNG plots per festival with period-based colors
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    df['revenue'] = df['final_amount_inr']
    
    all_fest_data = []

    festivals = df['festival_name'].dropna().unique()
    
    for fest in festivals:
        fest_data = df[df['festival_name'] == fest].copy()
        if fest_data.empty:
            continue
        
        fest_start = fest_data['order_date'].min()
        fest_end = fest_data['order_date'].max()
        
        # Define Before/During/After
        def fest_period(order_date):
            if order_date < fest_start:
                return 'Before'
            elif fest_start <= order_date <= fest_end:
                return 'During'
            else:
                return 'After'
        
        fest_data['period'] = fest_data['order_date'].apply(fest_period)
        fest_data['festival_name'] = fest
        all_fest_data.append(fest_data[['festival_name', 'order_date', 'period', 'revenue']])
        
        # Aggregate daily revenue
        daily_revenue = fest_data.groupby(['order_date', 'period'])['revenue'].sum().reset_index()
        
        # Plot with fixed colors
        color_map = {'Before':'blue', 'During':'red', 'After':'green'}
        fig = px.line(
            daily_revenue,
            x='order_date',
            y='revenue',
            color='period',
            color_discrete_map=color_map,
            title=f"Revenue Before/During/After: {fest}",
            labels={'revenue':'Revenue (INR)', 'order_date':'Date'}
        )
        fig.write_image(f"{save_path}/{fest}_revenue_time_series.png")
    
    # Combine all festivals for CSV
    combined_fest = pd.concat(all_fest_data, ignore_index=True)
    combined_fest.to_csv(f"{save_path}/all_festivals_daily_revenue.csv", index=False)
    
    print(f"Individual festival plots and combined CSV saved in '{save_path}'")
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import os
# question 10:
def price_demand_analysis_with_segments(df, save_path="output", segment_col='customer_tier'):
    """
    Price vs Demand Analysis by subcategory and customer segment.
    Saves:
        - Combined correlation CSV including customer segments
        - Scatter plots and correlation heatmaps per subcategory
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # Ensure numeric columns
    df = df[df['discounted_price_inr'].notna() & df['quantity'].notna()]
    df['revenue'] = df['final_amount_inr']
    
    subcategories = df['subcategory'].dropna().unique()
    
    correlation_summary = []

    for subcat in subcategories:
        subcat_data = df[df['subcategory'] == subcat].copy()
        if subcat_data.empty:
            continue
        
        # --- Scatter Plot: Price vs Quantity ---
        fig = px.scatter(
            subcat_data,
            x='discounted_price_inr',
            y='quantity',
            color=segment_col,
            hover_data=['category', 'brand', 'customer_age_group'],
            title=f"Price vs Demand: {subcat}",
            labels={'discounted_price_inr':'Discounted Price (INR)', 'quantity':'Quantity Sold'}
        )
        fig.write_image(f"{save_path}/scatter_price_vs_demand_{subcat}.png")
        
        # --- Correlation by Segment ---
        segments = subcat_data[segment_col].dropna().unique()
        for seg in segments:
            seg_data = subcat_data[subcat_data[segment_col] == seg]
            if seg_data.empty:
                continue
            
            corr_cols = ['discounted_price_inr', 'quantity', 'revenue']
            corr_matrix = seg_data[corr_cols].corr()
            
            correlation_summary.append({
                'subcategory': subcat,
                'customer_segment': seg,
                'avg_price': seg_data['discounted_price_inr'].mean(),
                'total_quantity': seg_data['quantity'].sum(),
                'total_revenue': seg_data['revenue'].sum(),
                'price_quantity_corr': corr_matrix.loc['discounted_price_inr','quantity'],
                'price_revenue_corr': corr_matrix.loc['discounted_price_inr','revenue'],
                'quantity_revenue_corr': corr_matrix.loc['quantity','revenue']
            })
            
            # --- Correlation heatmap per segment ---
            plt.figure(figsize=(6,4))
            sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
            plt.title(f"Correlation Matrix: {subcat} | {seg}")
            plt.tight_layout()
            plt.savefig(f"{save_path}/corr_matrix_{subcat}_{seg}.png")
            plt.close()
    
    # Save combined correlation summary
    pd.DataFrame(correlation_summary).to_csv(f"{save_path}/price_quantity_correlation_summary_by_segment.csv", index=False)
    
    print(f"Price vs demand analysis completed. Scatter plots, correlation heatmaps, and combined CSV with segments saved in '{save_path}'")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
# question 12:
def return_analysis(df, save_path="output"):
    """
    Analyze return patterns and customer satisfaction.
    Saves:
        - Summary CSV for returns and correlations
        - Plots as PNGs
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # Ensure relevant columns
    df = df[df['return_status'].notna() & df['product_rating'].notna() & df['discounted_price_inr'].notna()]
    
    # --- 1. Return rates by subcategory ---
    df['is_returned'] = df['return_status'].str.lower().eq('returned')
    
    summary = df.groupby(['subcategory','customer_tier','customer_city']).agg(
        total_orders=('transaction_id','count'),
        total_returns=('is_returned','sum')
    ).reset_index()
    
    summary['return_rate'] = summary['total_returns'] / summary['total_orders']
    
    # --- 2. Correlation with product rating and price ---
    corr_summary = df.groupby(['subcategory','customer_tier','customer_city']).apply(
        lambda x: pd.Series({
            'return_rating_corr': x['is_returned'].corr(x['product_rating']),
            'return_price_corr': x['is_returned'].corr(x['discounted_price_inr'])
        })
    ).reset_index()
    
    final_summary = pd.merge(summary, corr_summary, on=['subcategory','customer_tier','customer_city'])
    final_summary.to_csv(f"{save_path}/return_patterns_summary.csv", index=False)
    
    # --- 3. Return rate barplot by subcategory ---
    plt.figure(figsize=(14,6))
    sns.barplot(x='subcategory', y='return_rate', hue='customer_tier', data=final_summary)
    plt.xticks(rotation=45, ha='right')
    plt.title("Return Rate by Subcategory and Customer Tier")
    plt.tight_layout()
    plt.savefig(f"{save_path}/return_rate_subcategory.png")
    plt.close()
    
    # --- 4. Correlation heatmap (returns vs rating/price) ---
    # Aggregate over cities to avoid duplicates
    heatmap_data_rating = final_summary.groupby(['subcategory','customer_tier'])['return_rating_corr'].mean().reset_index()
    pivot_rating = heatmap_data_rating.pivot(index='subcategory', columns='customer_tier', values='return_rating_corr')

    plt.figure(figsize=(12,8))
    sns.heatmap(pivot_rating, annot=True, cmap='coolwarm', center=0)
    plt.title("Correlation: Return vs Product Rating")
    plt.tight_layout()
    plt.savefig(f"{save_path}/return_rating_corr_heatmap.png")
    plt.close()

    heatmap_data_price = final_summary.groupby(['subcategory','customer_tier'])['return_price_corr'].mean().reset_index()
    pivot_price = heatmap_data_price.pivot(index='subcategory', columns='customer_tier', values='return_price_corr')

    plt.figure(figsize=(12,8))
    sns.heatmap(pivot_price, annot=True, cmap='coolwarm', center=0)
    plt.title("Correlation: Return vs Discounted Price")
    plt.tight_layout()
    plt.savefig(f"{save_path}/return_price_corr_heatmap.png")
    plt.close()

    print(f"Return analysis completed. Summary CSV and plots saved in '{save_path}'")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
# question 13:
def brand_market_analysis(df, save_path="output"):
    """
    Analyze brand performance and market share.
    Saves:
        - Summary CSV for revenue, market share, and product count
        - Plots (PNG) for trends and comparison
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # Ensure relevant columns
    df = df[df['final_amount_inr'].notna() & df['brand'].notna() & df['subcategory'].notna()]
    
    # --- 1. Summary: revenue and market share per brand per subcategory ---
    summary = df.groupby(['subcategory','brand']).agg(
        total_revenue=('final_amount_inr','sum'),
        total_orders=('transaction_id','count'),
        num_products=('product_id','nunique')
    ).reset_index()
    
    # Compute market share per subcategory

    summary['market_share'] = summary.groupby('subcategory')['total_revenue'].transform(lambda x: x / x.sum())

    
    summary.to_csv(f"{save_path}/brand_market_summary.csv", index=False)
    
    # --- 2. Brand revenue comparison barplots per subcategory ---
    for subcat in summary['subcategory'].unique():
        sub = summary[summary['subcategory']==subcat]
        plt.figure(figsize=(12,6))
        sns.barplot(x='brand', y='total_revenue', data=sub.sort_values('total_revenue', ascending=False))
        plt.xticks(rotation=45, ha='right')
        plt.title(f"Brand Revenue Comparison - {subcat}")
        plt.tight_layout()
        plt.savefig(f"{save_path}/brand_revenue_{subcat}.png")
        plt.close()
    
    # --- 3. Market share pie chart per subcategory ---
    for subcat in summary['subcategory'].unique():
        sub = summary[summary['subcategory']==subcat]
        plt.figure(figsize=(8,8))
        plt.pie(sub['market_share'], labels=sub['brand'], autopct='%1.1f%%', startangle=140)
        plt.title(f"Market Share by Brand - {subcat}")
        plt.tight_layout()
        plt.savefig(f"{save_path}/brand_marketshare_{subcat}.png")
        plt.close()
    
    # --- 4. Revenue trend over years per brand (if order_year available) ---
    if 'order_year' in df.columns:
        trend = df.groupby(['subcategory','brand','order_year']).agg(
            yearly_revenue=('final_amount_inr','sum')
        ).reset_index()
        
        for subcat in trend['subcategory'].unique():
            sub = trend[trend['subcategory']==subcat]
            plt.figure(figsize=(12,6))
            sns.lineplot(x='order_year', y='yearly_revenue', hue='brand', data=sub, marker='o')
            plt.title(f"Brand Revenue Trend Over Years - {subcat}")
            plt.tight_layout()
            plt.savefig(f"{save_path}/brand_trend_{subcat}.png")
            plt.close()
    
    print(f"Brand and market share analysis completed. CSV and plots saved in '{save_path}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# question 14:
def clv_cohort_analysis(df, save_path="output"):
    """
    Customer Lifetime Value (CLV) & Cohort Analysis
    Saves:
        - Summary CSV for CLV per cohort
        - Cohort retention heatmap
        - Retention curves
        - CLV distribution plots
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # Ensure relevant columns
    df = df[df['order_date'].notna() & df['customer_id'].notna() & df['final_amount_inr'].notna()]
    
    # Convert order_date to datetime
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    df = df[df['order_date'].notna()]
    
    # --- Step 1: Create acquisition cohort (month of first purchase) ---
    df['order_month'] = df['order_date'].dt.to_period('M')
    df['cohort_month'] = df.groupby('customer_id')['order_month'].transform('min')
    
    # --- Step 2: Calculate months since cohort ---
    df['cohort_index'] = (df['order_month'].dt.year - df['cohort_month'].dt.year) * 12 + \
                         (df['order_month'].dt.month - df['cohort_month'].dt.month) + 1
    
    # --- Step 3: Cohort retention table ---
    cohort_data = df.groupby(['cohort_month','cohort_index'])['customer_id'].nunique().reset_index()
    cohort_counts = cohort_data.pivot(index='cohort_month', columns='cohort_index', values='customer_id')
    
    cohort_size = cohort_counts.iloc[:,0]
    retention = cohort_counts.divide(cohort_size, axis=0)
    
    # --- Step 4: Plot cohort retention heatmap ---
    plt.figure(figsize=(12,8))
    sns.heatmap(retention, annot=True, fmt=".0%", cmap='YlGnBu')
    plt.title("Cohort Retention Heatmap")
    plt.ylabel("Cohort Month")
    plt.xlabel("Months Since First Purchase")
    plt.tight_layout()
    plt.savefig(f"{save_path}/cohort_retention_heatmap.png")
    plt.close()
    
    # --- Step 5: Calculate CLV per customer ---
    clv = df.groupby('customer_id')['final_amount_inr'].sum().reset_index()
    clv_summary = clv.merge(df[['customer_id','customer_tier','order_year']].drop_duplicates(), on='customer_id', how='left')
    clv_summary.to_csv(f"{save_path}/customer_lifetime_value_summary.csv", index=False)
    
    # --- Step 6: Plot CLV distribution ---
    plt.figure(figsize=(12,6))
    sns.histplot(clv_summary['final_amount_inr'], bins=50, kde=True, color='skyblue')
    plt.title("Customer Lifetime Value Distribution")
    plt.xlabel("CLV (INR)")
    plt.ylabel("Number of Customers")
    plt.tight_layout()
    plt.savefig(f"{save_path}/clv_distribution.png")
    plt.close()
    
    # --- Step 7: CLV by tier and acquisition year ---
    plt.figure(figsize=(12,6))
    sns.boxplot(x='customer_tier', y='final_amount_inr', hue='order_year', data=clv_summary)
    plt.title("CLV by Customer Tier and Acquisition Year")
    plt.ylabel("CLV (INR)")
    plt.xlabel("Customer Tier")
    plt.legend(title='Acquisition Year')
    plt.tight_layout()
    plt.savefig(f"{save_path}/clv_tier_year.png")
    plt.close()
    
    print(f"CLV & Cohort analysis completed. Summary CSV and plots saved in '{save_path}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# question 16:

def product_rating_sales_analysis(df, save_path):
    """
    Analyze product rating patterns and their impact on sales.
    Includes:
      - Rating distribution
      - Correlation between rating, price, and sales
      - Subcategory-level insights
    Saves summary CSV and visualizations as PNG.
    """
    os.makedirs(save_path, exist_ok=True)

    # --- Step 1: Basic cleaning ---
    if 'product_rating' not in df.columns or 'final_amount_inr' not in df.columns:
        raise ValueError("Required columns: 'product_rating', 'final_amount_inr'")

    # Ensure numeric types
    df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')
    df['final_amount_inr'] = pd.to_numeric(df['final_amount_inr'], errors='coerce')
    df['product_price_inr'] = pd.to_numeric(df.get('product_price_inr', pd.NA), errors='coerce')
    df['quantity'] = pd.to_numeric(df.get('quantity', 1), errors='coerce')

    # Drop missing
    df = df.dropna(subset=['product_rating', 'final_amount_inr'])

    # --- Step 2: Subcategory level aggregation ---
    summary = df.groupby('subcategory').agg(
        avg_rating=('product_rating', 'mean'),
        total_sales=('final_amount_inr', 'sum'),
        avg_price=('product_price_inr', 'mean'),
        total_quantity=('quantity', 'sum')
    ).reset_index()

    # --- Step 3: Correlations ---
    summary['rating_sales_corr'] = df.groupby('subcategory').apply(
        lambda x: x['product_rating'].corr(x['final_amount_inr'])
    ).values
    summary['rating_price_corr'] = df.groupby('subcategory').apply(
        lambda x: x['product_rating'].corr(x['product_price_inr'])
    ).values

    # Save summary
    summary.to_csv(f"{save_path}/product_rating_sales_summary.csv", index=False)

    # --- Step 4: Rating Distribution ---
    plt.figure(figsize=(10,6))
    sns.histplot(df['product_rating'], bins=20, kde=True, color='teal')
    plt.title("Product Rating Distribution")
    plt.xlabel("Product Rating")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig(f"{save_path}/rating_distribution.png")
    plt.close()

    # --- Step 5: Rating vs Sales Scatter ---
    plt.figure(figsize=(10,6))
    sns.scatterplot(data=df, x='product_rating', y='final_amount_inr', hue='subcategory', alpha=0.7)
    plt.title("Rating vs Sales Amount by Subcategory")
    plt.xlabel("Product Rating")
    plt.ylabel("Final Amount (INR)")
    plt.tight_layout()
    plt.savefig(f"{save_path}/rating_vs_sales_scatter.png")
    plt.close()

    # --- Step 6: Rating vs Price correlation heatmap ---
    plt.figure(figsize=(10,6))
    corr_matrix = summary[['avg_rating','avg_price','total_sales','total_quantity']].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title("Correlation Between Rating, Price, Sales & Quantity")
    plt.tight_layout()
    plt.savefig(f"{save_path}/rating_price_sales_correlation_heatmap.png")
    plt.close()

    print(f"✅ Rating analysis completed! Saved summary & charts in: {save_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import os
# question 17:
def customer_journey_analysis(df, save_path):
    """
    Customer Journey Analysis:
    - Purchase frequency & loyalty segments
    - Category transition matrix
    - Sankey flow diagram
    - Customer evolution summary
    Saves: CSVs, PNGs, and HTML visualizations.
    """
    
    os.makedirs(save_path, exist_ok=True)
    
    # --- Step 1: Data Preparation ---
    required_cols = ['customer_id', 'transaction_id', 'order_date', 'subcategory']
    missing_cols = [c for c in required_cols if c not in df.columns]
    if missing_cols:
        raise KeyError(f"Missing columns: {missing_cols}")
    
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    df = df.dropna(subset=['order_date'])
    df = df.sort_values(['customer_id', 'order_date'])
    
    # --- Step 2: Purchase Frequency Analysis ---
    purchase_counts = df.groupby('customer_id')['transaction_id'].nunique().reset_index()
    purchase_counts.columns = ['customer_id', 'purchase_count']
    
    # Segment customers
    purchase_counts['segment'] = pd.cut(
        purchase_counts['purchase_count'],
        bins=[0, 1, 3, 6, np.inf],
        labels=['New', 'Occasional', 'Frequent', 'Loyal']
    )
    
    plt.figure(figsize=(8,5))
    sns.countplot(data=purchase_counts, x='segment', palette='viridis', order=['New','Occasional','Frequent','Loyal'])
    plt.title("Customer Segment Distribution (Based on Purchase Frequency)")
    plt.xlabel("Customer Segment")
    plt.ylabel("Number of Customers")
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, "customer_segment_distribution.png"))
    plt.close()
    
    # --- Step 3: Subcategory Transition Matrix ---
    df['next_subcategory'] = df.groupby('customer_id')['subcategory'].shift(-1)
    transitions = df.dropna(subset=['next_subcategory'])
    
    transition_matrix = pd.crosstab(
        transitions['subcategory'],
        transitions['next_subcategory'],
        normalize='index'
    ).round(3)
    
    plt.figure(figsize=(10,8))
    sns.heatmap(transition_matrix, annot=True, fmt=".2f", cmap='YlGnBu')
    plt.title("Category Transition Probability Matrix")
    plt.xlabel("Next Category")
    plt.ylabel("Current Category")
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, "category_transition_matrix.png"))
    plt.close()
    
    transition_matrix.to_csv(os.path.join(save_path, "category_transition_matrix.csv"))
    
    # --- Step 4: Sankey Flow Diagram ---
    top_transitions = (
        transitions.groupby(['subcategory', 'next_subcategory'])
        .size()
        .reset_index(name='count')
        .sort_values('count', ascending=False)
        .head(40)
    )

    labels = list(pd.unique(top_transitions[['subcategory', 'next_subcategory']].values.ravel('K')))
    label_to_index = {label: i for i, label in enumerate(labels)}

    sources = top_transitions['subcategory'].map(label_to_index)
    targets = top_transitions['next_subcategory'].map(label_to_index)
    values = top_transitions['count']
    
    fig = go.Figure(data=[go.Sankey(
        arrangement="snap",
        node=dict(
            pad=20,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=labels,
            color="rgba(66, 135, 245, 0.7)"
        ),
        link=dict(
            source=sources,
            target=targets,
            value=values,
            color="rgba(102, 194, 165, 0.5)"
        )
    )])
    
    fig.update_layout(
        title_text="Customer Journey: Category Transition Flow",
        font_size=11,
        paper_bgcolor="white"
    )
    
    sankey_path = os.path.join(save_path, "customer_journey_flow.html")
    fig.write_html(sankey_path)
    
    # --- Step 5: Customer Evolution Over Time ---
    df['order_year'] = df['order_date'].dt.year
    evolution = df.groupby(['order_year', 'customer_id']).size().reset_index(name='purchases')
    evolution_summary = evolution.groupby('order_year')['purchases'].mean().reset_index()
    
    plt.figure(figsize=(8,5))
    plt.plot(evolution_summary['order_year'], evolution_summary['purchases'], marker='o')
    plt.title("Average Purchases per Customer Over Time")
    plt.xlabel("Year")
    plt.ylabel("Average Purchases")
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, "customer_evolution_trend.png"))
    plt.close()
    
    # --- Step 6: Save summary CSV ---
    purchase_counts.to_csv(os.path.join(save_path, "customer_segments.csv"), index=False)
    
    print("✅ Customer Journey Analysis Completed Successfully!")
    print(f"✔ Sankey Diagram: {sankey_path}")
    print(f"✔ Results saved in: {save_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
#question 18:
def product_lifecycle_analysis(sales_csv, catalog_csv, save_path):
    """
    Analyze product lifecycle and revenue trends.
    
    Args:
        sales_csv (str): Path to cleaned Amazon sales CSV (2015-2025)
        catalog_csv (str): Path to product catalog CSV
        save_path (str): Folder to save results (CSVs + PNGs)
    """
    
    os.makedirs(save_path, exist_ok=True)
    
    # --- Load datasets ---
    df_sales = pd.read_csv(sales_csv)
    df_catalog = pd.read_csv(catalog_csv)
    
    # Ensure order_date is datetime
    df_sales['order_date'] = pd.to_datetime(df_sales['order_date'], errors='coerce')
    
    # --- Merge catalog + sales ---
    df = df_catalog.merge(
        df_sales[['product_id', 'order_date', 'final_amount_inr']],
        on='product_id', how='left'
    )
    
    # Drop rows without order_date (if any)
    df = df.dropna(subset=['order_date'])
    
    # --- Compute product age in months ---
    df['product_age_months'] = (df['order_date'].dt.year - df['launch_year']) * 12 + df['order_date'].dt.month
    
    # --- Aggregate monthly revenue per product ---
    df['revenue'] = df['final_amount_inr']
    product_monthly = df.groupby(
        ['product_id','product_name','subcategory','launch_year','product_age_months'],
        as_index=False
    )['revenue'].sum()
    
    # --- Determine peak revenue month and value per product ---
    peak_info = product_monthly.groupby('product_id').apply(
        lambda x: pd.Series({
            'product_name': x['product_name'].iloc[0],
            'subcategory': x['subcategory'].iloc[0],
            'launch_year': x['launch_year'].iloc[0],
            'peak_revenue': x['revenue'].max(),
            'peak_age_months': x.loc[x['revenue'].idxmax(), 'product_age_months']
        })
    ).reset_index()
    
    # --- Identify lifecycle phases ---
    def lifecycle_phase(row):
        if row['product_age_months'] < row['peak_age_months']:
            return 'Growth'
        elif row['product_age_months'] == row['peak_age_months']:
            return 'Peak'
        else:
            return 'Decline'
    
    product_monthly = product_monthly.merge(peak_info[['product_id','peak_age_months']], on='product_id', how='left')
    product_monthly['phase'] = product_monthly.apply(lifecycle_phase, axis=1)
    
    # --- Save summary CSV ---
    product_monthly.to_csv(os.path.join(save_path, 'product_lifecycle_summary.csv'), index=False)
    
    # --- Plot revenue trends per subcategory ---
    import seaborn as sns

# --- Plot revenue trends per subcategory (clearer) ---
    top_n = 5  # Number of products to show per subcategory
    subcategories = product_monthly['subcategory'].unique()

    for subcat in subcategories:
        df_sub = product_monthly[product_monthly['subcategory'] == subcat]
        
        # Pick top N products by total revenue
        top_products = df_sub.groupby('product_name')['revenue'].sum().nlargest(top_n).index
        df_top = df_sub[df_sub['product_name'].isin(top_products)]
        
        plt.figure(figsize=(12,6))
        sns.lineplot(
            data=df_top, 
            x='product_age_months', 
            y='revenue', 
            hue='product_name', 
            style='phase', 
            markers=True, 
            dashes=False, 
            palette='tab10'
        )
        
        plt.title(f"Top {top_n} Product Revenue Lifecycle - {subcat}")
        plt.xlabel("Product Age (Months)")
        plt.ylabel("Revenue (INR)")
        plt.grid(True)
        plt.legend(title='Product / Phase', fontsize='small', ncol=2)
        plt.tight_layout()
        plt.savefig(os.path.join(save_path, f"{subcat}_lifecycle.png"))
        plt.close()

    
    print(f"Product lifecycle analysis completed. CSV + plots saved in: {save_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
#question 19:
def competitive_pricing_analysis(
    sales_csv,
    save_path
):
    os.makedirs(save_path, exist_ok=True)

    # --- Load datasets ---
    df_sales = pd.read_csv(sales_csv)

    # --- Use sales data (already has subcategory, brand) ---
    df = df_sales

    # --- Compute average selling price per product ---
    df['selling_price'] = df['final_amount_inr'] / df['quantity']

    # --- Aggregate price info per brand/subcategory ---
    brand_pricing = df.groupby(['subcategory','brand']).agg(
        min_price=('selling_price','min'),
        max_price=('selling_price','max'),
        median_price=('selling_price','median'),
        mean_price=('selling_price','mean'),
        total_units=('quantity','sum'),
        total_revenue=('final_amount_inr','sum')
    ).reset_index()

    # --- Save summary CSV ---
    brand_pricing.to_csv(os.path.join(save_path, "brand_pricing_summary.csv"), index=False)

    # --- Get list of subcategories ---
    subcategories = df['subcategory'].unique()

    # --- Plot boxplots for price distributions per subcategory ---
    for subcat in subcategories:
        df_sub = df[df['subcategory'] == subcat]
        plt.figure(figsize=(12,6))
        sns.boxplot(x='brand', y='selling_price', data=df_sub, palette='tab20')
        plt.title(f"Price Distribution by Brand - {subcat}")
        plt.xlabel("Brand")
        plt.ylabel("Selling Price (INR)")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(os.path.join(save_path, f"{subcat}_pricing_boxplot.png"))
        plt.close()

    # --- Competitive Matrix: Price Tier vs Market Penetration ---
    # Compute price tiers based on median selling price
    brand_pricing['price_tier'] = pd.qcut(brand_pricing['median_price'], 3, labels=['Low','Mid','High'])

    for subcat in subcategories:
        df_sub = brand_pricing[brand_pricing['subcategory'] == subcat]
        plt.figure(figsize=(10,6))
        # Bubble size proportional to total_units, color by brand
        scatter = plt.scatter(
            x=df_sub['price_tier'],
            y=df_sub['total_units'],
            s=df_sub['total_units']/df_sub['total_units'].max()*500,
            c=pd.factorize(df_sub['brand'])[0],
            cmap='tab20',
            alpha=0.7
        )
        # Annotate brand names
        for i, row in df_sub.iterrows():
            plt.text(row['price_tier'], row['total_units'], row['brand'], fontsize=9, ha='center', va='bottom')
        plt.title(f"Competitive Matrix - {subcat}")
        plt.xlabel("Price Tier")
        plt.ylabel("Total Units Sold")
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.savefig(os.path.join(save_path, f"{subcat}_competitive_matrix.png"))
        plt.close()

    print("Competitive pricing analysis completed successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
#question 20:
def business_health_dashboard(
    sales_csv,
    save_path
):
    os.makedirs(save_path, exist_ok=True)

    # --- Load sales data ---
    df = pd.read_csv(sales_csv)
    df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
    df = df.dropna(subset=['order_date'])

    # --- Extract year and month ---
    df['year'] = df['order_date'].dt.year
    df['month'] = df['order_date'].dt.month

    # --- 1. Revenue growth per year ---
    revenue_yearly = df.groupby('year')['final_amount_inr'].sum().reset_index()
    revenue_yearly.rename(columns={'final_amount_inr':'value'}, inplace=True)
    revenue_yearly['metric'] = 'revenue_growth'

    # --- 2. Customer acquisition per year ---
    first_purchase = df.groupby('customer_id')['order_date'].min().reset_index()
    first_purchase['year'] = first_purchase['order_date'].dt.year
    new_customers = first_purchase.groupby('year')['customer_id'].nunique().reset_index()
    new_customers.rename(columns={'customer_id':'value'}, inplace=True)
    new_customers['metric'] = 'new_customers'

    # --- 3. Retention rate per year ---
    retention = []
    years = sorted(df['year'].unique())
    for i in range(1, len(years)):
        prev_year = years[i-1]
        curr_year = years[i]
        prev_customers = set(df[df['year']==prev_year]['customer_id'].unique())
        curr_customers = set(df[df['year']==curr_year]['customer_id'].unique())
        retained = len(prev_customers & curr_customers)
        retention_rate = retained / len(prev_customers) if len(prev_customers)>0 else 0
        retention.append({'year': curr_year, 'value': retention_rate, 'metric': 'retention_rate'})
    retention_df = pd.DataFrame(retention)

    # --- 4. Operational efficiency: avg delivery days per year ---
    delivery_efficiency = df.groupby('year')['delivery_days'].mean().reset_index()
    delivery_efficiency.rename(columns={'delivery_days':'value'}, inplace=True)
    delivery_efficiency['metric'] = 'avg_delivery_days'

    # --- Combine all metrics into one CSV ---
    combined_df = pd.concat([revenue_yearly, new_customers, retention_df, delivery_efficiency], ignore_index=True)
    combined_df.to_csv(os.path.join(save_path, "business_health_summary.csv"), index=False)

    # --- Plot multi-panel dashboard ---
    plt.figure(figsize=(16,12))

    # Revenue growth
    plt.subplot(2,2,1)
    sns.barplot(x='year', y='value', data=revenue_yearly, palette='Blues_d')
    plt.title("Revenue Growth")
    plt.ylabel("Revenue (INR)")

    # New customers
    plt.subplot(2,2,2)
    sns.barplot(x='year', y='value', data=new_customers, palette='Greens_d')
    plt.title("New Customers Acquired")
    plt.ylabel("Customers")

    # Retention rate
    plt.subplot(2,2,3)
    sns.lineplot(x='year', y='value', data=retention_df, marker='o', color='orange')
    plt.title("Customer Retention Rate")
    plt.ylabel("Retention Rate")

    # Avg delivery days
    plt.subplot(2,2,4)
    sns.lineplot(x='year', y='value', data=delivery_efficiency, marker='o', color='red')
    plt.title("Avg Delivery Days")
    plt.ylabel("Days")

    plt.tight_layout()
    plt.savefig(os.path.join(save_path, "business_health_dashboard.png"))
    plt.close()

    print("Business health dashboard completed! CSV and PNG saved successfully.")
```
